```python
from random import sample
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def dot(v1, v2):
    r = 0
    for i in range(len(v1)):
        r += v1[i] * v2[i]
    return r

def PLA(data):
    w = np.array([0,0]) #(w1,w2), w0=0
    run = True
    while run:
        run = False
        for x in data:
            if dot(w, x[0:2])*x[2] <= 0:
                w +=  x[0:2]*x[2]
                run = True
            else:
                pass
    return w

def gen_pic(num, show=False):
    data_set = []
    for i in range(30):
        x1, x2 = sample(range(-25,26,3), 2)
        y = 0
        if x1 > x2:
            y = 1
        else:
            y = -1
        data_set.append(np.array([x1,x2,y]))

    wf = PLA(data_set)
    
    pd_data = pd.DataFrame(data_set, columns = ["x1", "x2", "y"])
    #print(pd_data)
    
    my_dpi = 250
    plt.figure(figsize=(1000/my_dpi, 1000/my_dpi), dpi=my_dpi)
    
    g = pd_data.loc[pd_data['y'] == 1]
    plt.scatter(g["x1"], g["x2"], marker="o", color="blue")

    b = pd_data.loc[pd_data['y'] == -1]
    plt.scatter(b["x1"], b["x2"], marker="x", color="red")

    x = np.linspace(-30,30,1000)
    plt.plot(-wf[1]*x, wf[0]*x, c="purple")

    plt.xlim(-30, 30)
    plt.ylim(-30, 30)
    plt.title(f"PLA found line {wf[0]}x+{wf[1]}y=0")
    
    plt.savefig(f'number {num}.png', bbox_inches='tight', dpi=my_dpi)
    if show:
        plt.show()
    plt.clf()


# data_set = []
# for i in range(10):
#     x1, x2 = sample(range(-25,26,2), 2)
#     y = 0
#     if x1 > x2:
#         y = 1
#     else:
#         y = -1
#     data_set.append(np.array([x1,x2,y]))

# ##############################################################
# wf = PLA(data_set)
# print(f"{wf[0]}x+{wf[1]}y=0")
# ##############################################################
# pd_data = pd.DataFrame(data_set, columns = ["x1", "x2", "y"])
# #print(pd_data)

# g = pd_data.loc[pd_data['y'] == 1]
# plt.scatter(g["x1"], g["x2"], marker="o", color="blue")

# b = pd_data.loc[pd_data['y'] == -1]
# plt.scatter(b["x1"], b["x2"], marker="x", color="red")

# x = np.linspace(-30,30,1000)
# plt.plot(-wf[1]*x, wf[0]*x, c="purple")

# plt.xlim(-30, 30)
# plt.ylim(-30, 30)
# plt.title(f"PLA found line {wf[0]}x+{wf[1]}y=0")

# plt.savefig('foo.png', bbox_inches='tight')
# #plt.show()
```
